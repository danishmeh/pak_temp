```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df= pd.read_csv("Tempreture_1901_2016_Pakistan.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.rename(columns ={'Temperature - (Celsius)':'temp',' Year':'year','Month':'month'},inplace =True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
fig = px.scatter(df, x="year", y="temp", color="month")
fig.show()
```

```python
fig = px.line(df, x="year", y="temp", color="month")
fig.show()
```

```python
# January (116, 3) 
d_jan= df.query('year >=1901').query("month =='January'")
fig = px.line(d_jan, x="year", y="temp", title='Tempearture in January',markers=True
              )
fig.show()
```

## February

```python
df_feb= df.query('year >=1901').query("month =='February'")
fig = px.line(df_feb, x="year", y="temp", title='Tempearture in February ',markers=True,
            )
fig.update_traces(line_color="#32CD32")
fig.show()
```

```python
df_mar = df.query('year >=1901').query("month =='March'")
fig = px.line(df_mar, x="year", y="temp", title='Tempearture in March',markers=True,)
fig.update_traces(line_color="#32CD32")
fig.show()
```

```python
df_j = df.query('year >=1901').query("month =='June'")
fig = px.line(df_j, x="year", y="temp", title='Tempearture in June',markers=True,)
fig.update_traces(line_color="#32CD32")
fig.show()
```

```python
df_july = df.query('year >=1901').query("month =='July'")
fig = px.line(df_july, x="year", y="temp", title='Tempearture in July',markers=True,)
fig.update_traces(line_color="#32CD32")
fig.show()
```

```python
plt.figure(figsize=(15,5))
df = df.groupby('year').mean()
plt.plot(df.index, df['temp'])
plt.xlabel('year')
plt.ylabel('Temp in Avg')
plt.title('Temp Record 1901 to 2016', y=1.1)
# fig.update_traces(line_color="#32CD32")
plt.grid()
plt.show()
```

```python
df['mean_of_temp'] = df.groupby('year').mean()
```
